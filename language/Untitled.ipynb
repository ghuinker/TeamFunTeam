```python
import tensorflow as tf
from keras.models import Sequential
import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = Sequential()
model.add(keras.layers.Dense(units=64, activation='relu',input_dim=100))
model.add(keras.layers.Dense(units=2,activation='softmax'))
```

```python
sgd = keras.optimizers.SGD(lr=0.3)
model.compile(loss='categorical_crossentropy',
             optimizer = sgd,
             metrics=['accuracy'])
```

```python
def randArray(size):
    seeds = [np.random.uniform(0,1) for i in range(size)]
    rand = []
    for i in seeds:
        if i < 0.5:
            rand.append(0)
        else:
            rand.append(1)
    return rand
```

```python
def notRandArray(size):
    lin = np.sin(np.linspace(0,2,size))
    return lin
```

```python
def noisySinCosArrays(size,vecDim, noiseLevel):
    xret = []
    yret = []
    linspace = np.linspace(0,3,vecDim)
    for i in range(size):
        seed = np.random.uniform(0,1) < 0.5
        if(seed):
            newVec = np.sin(linspace)
            newVec = noise(newVec, noiseLevel)
            xret.append(newVec)
            yret.append([0,1])
        else:
            newVec = np.cos(linspace)
            newVec = noise(newVec, noiseLevel)
            xret.append(newVec)
            yret.append([1,0])
    return np.array(xret), np.array(yret)
    
```

```python
def noise(array, noisiness):
    noises = np.array([np.random.uniform(-noisiness, noisiness) for i in range(len(array))])
    return array + noises
```
